### Analyse positional encodings

```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns

from math import sin, cos
```

```python
n = 40 
d = 64

p = np.zeros([n, d])

for j in range(n): 
    for i in range(d): 
        if i%2 == 0:
            p[j, i] = sin(j/(10000**(i/d)))
        else: 
            p[j, i] = cos(j/(10000**((i-1)/d)))
            
t = np.linspace(0, n-1, n)

plt.figure(figsize=(16,8))
plt.plot(t, p)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.imshow(p.T, cmap='Blues')
plt.show()
```

```python
#calculate distances 

dist = np.zeros([n,n])

for i in range(n): 
    for j in range(n): 
        dist[i,j] = np.sqrt(np.sum((p[i,:]-p[j,:]).T*(p[i,:]-p[j,:])))
```

```python
import seaborn as sns
sns.heatmap(dist, square = True, cmap = 'Blues').invert_yaxis()
```
